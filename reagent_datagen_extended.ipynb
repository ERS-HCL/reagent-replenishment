```python
def get_protocol_in_ml(REAGENT_NAME_OR_TEST_NAME = "Not Implemented"):
    ret_val = 1 # default value
    if (REAGENT_NAME_OR_TEST_NAME == "Reagent-AFP"):
        ret_val = .5
    if (REAGENT_NAME_OR_TEST_NAME == "Reagent-Anti-HAViGG"):
        ret_val = 2    
    if (REAGENT_NAME_OR_TEST_NAME == "Reagent-Anti-TPO"):
        ret_val = 1    
    if (REAGENT_NAME_OR_TEST_NAME == "Reagent-Anti-CCP"):
        ret_val = 3   
    if (REAGENT_NAME_OR_TEST_NAME == "Reagent-BNP"):
        ret_val = 1   
    return(ret_val)
```

```python
def get_reagent_wastage(number_of_obs, pack_size):
    LST_REAGENT_WASTAGE = []
    lst_retain = []
    for i in range(number_of_obs//12): # 12 because every few days (2,3) in 12 days we want to have reagent wastage.
        retain_value = ((i*12) + random.choice([x for x in range(0, 12)]))
        lst_retain.append(retain_value)
    #print(lst_retain)
    
    for i in range(number_of_obs):
        ret = 0
        if (i in lst_retain):
            x = (round(np.random.uniform(.8, 1.0) * 100) * .001) # getting a number between 8-10%
            ret = round(x*pack_size)
        LST_REAGENT_WASTAGE.append(ret) 
    
    return(LST_REAGENT_WASTAGE)
```

```python
def adjust_reagent_wastage(LST_REAGENT_WASTAGE):
    
    for i in range(len(LST_REAGENT_WASTAGE)):
      if (LST_REAGENT_WASTAGE[i] > 0):
        total_reduced_val = 0
        original_reagent_wastage = LST_REAGENT_WASTAGE[i]
        # We have found wastage let's distribute this
        distribute_days = random.choice([2,3])
        indices = np.random.randint(-6, 6, distribute_days) # Getting indices to update.
        for ix in indices:
          index_to_change = i + ix
          if ((index_to_change) < 0): i - ix
          index_to_change = min(len(LST_REAGENT_WASTAGE) - 1, index_to_change)
          prob = min(.8,random.uniform(0, 1/distribute_days))
          change_val = prob * original_reagent_wastage
          change_val = round(min(change_val,7),1)
          total_reduced_val += change_val
          if (i != index_to_change):
            LST_REAGENT_WASTAGE[index_to_change] = LST_REAGENT_WASTAGE[index_to_change] + change_val
            LST_REAGENT_WASTAGE[i] -= change_val
        
    for i in range(len(LST_REAGENT_WASTAGE)): # enhancing the dead volume
      if (LST_REAGENT_WASTAGE[i] > 0):
        if (LST_REAGENT_WASTAGE[i] < 3):
          LST_REAGENT_WASTAGE[i] = LST_REAGENT_WASTAGE[i] + 3
      
    return(LST_REAGENT_WASTAGE)
```

```python

def get_repeated_test_runs(REAGENT_NAME = "Not Implemented"):
    return(round(np.random.uniform(0, 1) * 10) % 10)

def get_calibration_run(REAGENT_NAME = "Not Implemented"):
    # 10-Sep-2019: Calibration run should be a fixed number or zero. 
    # 24 Sep - Should be a fixed number (6 or 9) or zero
    # Generally every seven days , we put a value
    # equal to test runs.
    ret_val = 6
    if (REAGENT_NAME == "Reagent-AFP"):
        ret_val = 6
    if (REAGENT_NAME == "Reagent-Anti-HAViGG"):
        ret_val = 6    
    if (REAGENT_NAME == "Reagent-Anti-TPO"):
        ret_val = 9    
    if (REAGENT_NAME == "Reagent-Anti-CCP"):
        ret_val = 9   
    if (REAGENT_NAME == "Reagent-BNP"):
        ret_val = 9   
    return(ret_val)
def get_qc_failure(REAGENT_NAME = "Not Implemented"):
    return(random.choice([0,1,2,3,4,5]))
def get_test_runs(pack_size = 300):
    # returning a value between 8 and 12.
    return(round(pack_size * round(np.random.uniform(.8, 1.2) * 10) * .01))
```

```python
def get_date_list( start_date = '2017-01-01', num_days = 100):
    lst_dates = []
    start_date = datetime.datetime.strptime(start_date, '%Y-%m-%d')
    for i in range(num_days):
        next_date = start_date + datetime.timedelta(i)
        lst_dates.append(datetime.date.strftime(next_date, "%d-%m-%Y"))
    return(lst_dates)
```

```python
def get_total_reagent_volume(number_of_obs = 10, reagent_name = "Reagent-AFP", pack_size = 300):
    from operator import add
    LST_REAGENT_NAME = [reagent_name  for i in range(number_of_obs)]
    LST_PACK_SIZES = [pack_size  for i in range(number_of_obs)]
    LST_CALIBRATION_RUN = list(map(get_calibration_run, LST_REAGENT_NAME)) 
    LST_REPEATED_TEST_RUNS = list(map(get_repeated_test_runs, LST_REAGENT_NAME)) 
    LST_PROTOCOL_IN_ML = list(map(get_protocol_in_ml, LST_REAGENT_NAME)) 
    LST_QC_FAILURE_REPEAT = list(map(get_qc_failure_repeat, LST_REAGENT_NAME))
    LST_QC_CONTROL_RUN = list(map(get_qc_control_run, LST_REAGENT_NAME))
    LST_QC_FAILURE = list(map(get_qc_failure, LST_REAGENT_NAME))
    LST_TEST_RUNS = list(map(get_test_runs, LST_PACK_SIZES))
    return(LST_QC_FAILURE_REPEAT ,LST_PROTOCOL_IN_ML, LST_TEST_RUNS ,LST_QC_CONTROL_RUN, LST_CALIBRATION_RUN, LST_REPEATED_TEST_RUNS, LST_QC_FAILURE)
```

```python
def incr_list_values(LST_COLS, REAGENT_NAME = "Not Implemented", increment_by_percentage = 5, step_of_increment = 100, rounding = True):
    # Every 100 observation, increase in value is observed.
    adjustment_factor = increment_by_percentage
    
    if (REAGENT_NAME == "Reagent-AFP"):
        adjustment_factor = 7
    if (REAGENT_NAME == "Reagent-Anti-HAViGG"):
        adjustment_factor = 9    
    if (REAGENT_NAME == "Reagent-Anti-TPO"):
        adjustment_factor = 3    
    if (REAGENT_NAME == "Reagent-Anti-CCP"):
        adjustment_factor = 4   
    if (REAGENT_NAME == "Reagent-BNP"):
        adjustment_factor = 20  
    
    tmp_factor = 1
    for i in range(len(LST_COLS)):
        LST_COLS[i] = math.ceil(adjustment_factor * LST_COLS[i])
        if ((i+1) % step_of_increment == 0): adjustment_factor = adjustment_factor * (1 + (increment_by_percentage/100) )
    return(LST_COLS)
```

```python
def get_pack_increment(REAGENT_NAME = "Not Implemented"):
    ret_val = 3
    if (REAGENT_NAME == "Reagent-AFP"):
        ret_val = 3
    if (REAGENT_NAME == "Reagent-Anti-HAViGG"):
        ret_val = 3    
    if (REAGENT_NAME == "Reagent-Anti-TPO"):
        ret_val = 3    
    if (REAGENT_NAME == "Reagent-Anti-CCP"):
        ret_val = 3   
    if (REAGENT_NAME == "Reagent-BNP"):
        ret_val = 3   
    return(ret_val)

def get_reagent_increment(REAGENT_NAME = "Not Implemented"):
    # How much volume is added when one pack is opened.
    ret_val = 300
    if (REAGENT_NAME == "Reagent-AFP"):
        ret_val = 300
    if (REAGENT_NAME == "Reagent-Anti-HAViGG"):
        ret_val = 300    
    if (REAGENT_NAME == "Reagent-Anti-TPO"):
        ret_val = 300    
    if (REAGENT_NAME == "Reagent-Anti-CCP"):
        ret_val = 300   
    if (REAGENT_NAME == "Reagent-BNP"):
        ret_val = 300  
    return(ret_val)
```

```python
CONST_AV_REAGENT_VOL_PER_DAY_ANTI_HAVIGG = 60
CONST_AV_REAGENT_VOL_PER_DAY_AFP = 52
CONST_AV_REAGENT_VOL_PER_DAY_ANTI_CCP = 40

CONST_REAGENT_AFP_NAME = "Reagent-AFP"
CONST_REAGENT_ANTI_HAVIGG_NAME = "Reagent-Anti-HAViGG"
CONST_REAGENT_ANTI_CCP_NAME = "Reagent-Anti-CCP"

def get_reagent_wastage_old(REAGENT_NAME = "Not Implemented", NUMBER_OBS = 10):
    # Since we want this to 5% - 10% of total wastage, we can do with a normal distribution of mean
    # equalling to constants defined above say about 7.5%
    mu = 0
    sigma = 1
    s = np.random.normal(mu, sigma, NUMBER_OBS) # Default
    if (REAGENT_NAME == CONST_REAGENT_ANTI_HAVIGG_NAME):
        mu = CONST_AV_REAGENT_VOL_PER_DAY_ANTI_HAVIGG * 7.5 / 100
        sigma = 5
        s = np.random.normal(mu, sigma, NUMBER_OBS)
    if (REAGENT_NAME == CONST_REAGENT_AFP_NAME):
            mu = CONST_AV_REAGENT_VOL_PER_DAY_AFP * 12 / 100
            sigma = 10
            s = np.random.normal(mu, sigma, NUMBER_OBS)
    if (REAGENT_NAME == CONST_REAGENT_ANTI_CCP_NAME):
            mu = CONST_AV_REAGENT_VOL_PER_DAY_ANTI_CCP * 5 / 100
            sigma = 3
            s = np.random.normal(mu, sigma, NUMBER_OBS)
    x = abs(s)
    xy = [round(x) for x in x]
    return(xy)
```

```python
def get_data(number_of_obs = 2,
             pack_size = 300,
             ml_per_run = 2,
             reagent_name = "Reagent-AFP",
             initial_pack_units = 1,
             start_date = "2010-01-01",
             test_name = "Test_Default"):
    # -----------------------------------------------------------------------------------------
    # incr_reagent_inventory - What is the value with which 1 pack is replenished.
    # incr_packs - The number of packs with which when packs run low.
    # pack_size - is a constant.
    # number_of_runs - is number of observations.
    # -----------------------------------------------------------------------------------------
    incr_packs = get_pack_increment(reagent_name)
    incr_reagent_inventory = get_reagent_increment(reagent_name)
    LST_REAGENT_NAME = [reagent_name  for i in range(number_of_obs)]
    LST_DATES = get_date_list(num_days = number_of_obs, start_date = start_date)
    prev_volume  = pack_size * initial_pack_units
    tmp_pack_units = initial_pack_units
    LST_REAGENT_WORK_IN_PROGRESS_VOLUME = []
    LST_NUMBER_OF_PACK_UNITS = []
    LST_PACK_SIZE = [] # This will be constant for now.
    LST_WORK_STOCK_VOLUME = []
    LST_QC_TOTAL_RUNS = []
    tmp_reag_wip_vol = pack_size * initial_pack_units # RITESH
    # Getting the various constituents here
    LST_QC_FAILURE_REPEAT, LST_PROTOCOL_IN_ML, LST_TEST_RUNS,LST_QC_CONTROL_RUN,LST_CALIBRATION_RUN_SCHEDULED,LST_REPEATED_TEST_RUNS, LST_QC_FAILURE = get_total_reagent_volume(number_of_obs = number_of_obs, reagent_name = reagent_name, pack_size = pack_size)

    LST_CALIBRATION_RUN_ON_FAILURE = []
    for i in range(len(LST_CALIBRATION_RUN_SCHEDULED)):
        tmp_calibration_run_failure = 0
        if (i%7 != 0):
            LST_CALIBRATION_RUN_SCHEDULED[i] = 0
        if (LST_QC_FAILURE_REPEAT[i] > 0):
            tmp_calibration_run_failure = get_calibration_run(reagent_name)
        LST_CALIBRATION_RUN_ON_FAILURE.append(tmp_calibration_run_failure)
  
    LST_REAGENT_WASTAGE = get_reagent_wastage(number_of_obs, pack_size)
    LST_REAGENT_WASTAGE = adjust_reagent_wastage(LST_REAGENT_WASTAGE)
    # ----------------------------------------------------------------------------#
    LST_TOTAL =  list( map(add, LST_TEST_RUNS, LST_REPEATED_TEST_RUNS)) # Replaced LST_TOTAL by LST_TEST_RUNS
    LST_TOTAL =  list( map(add, LST_TOTAL, LST_REAGENT_WASTAGE))


    for i in range(number_of_obs):
        if (LST_QC_FAILURE[i] > LST_QC_CONTROL_RUN[i]):
            LST_QC_FAILURE[i] = 0
            
        if (LST_QC_FAILURE_REPEAT[i] > LST_QC_FAILURE[i]):
            LST_QC_FAILURE_REPEAT[i] = 0
            LST_CALIBRATION_RUN_ON_FAILURE[i] = 0
        temp_qc_tot_run = LST_QC_CONTROL_RUN[i] + LST_QC_FAILURE[i] + LST_QC_FAILURE_REPEAT[i]
        LST_QC_TOTAL_RUNS.append(temp_qc_tot_run)
      
    LST_CALIBRATION_RUN = list(map(add,LST_CALIBRATION_RUN_SCHEDULED, LST_CALIBRATION_RUN_ON_FAILURE))
    LST_TOTAL =  list( map(add, LST_TOTAL, LST_QC_TOTAL_RUNS))
    LST_TOTAL_REAGENT_VOLUME =  list( map(add, LST_TOTAL, LST_CALIBRATION_RUN)) 
    LST_TOTAL_REAGENT_VOLUME = np.multiply(LST_TOTAL_REAGENT_VOLUME, LST_PROTOCOL_IN_ML)
    
    for i in range(number_of_obs):
            # QC TOTAL RUNS
            deduct = LST_TOTAL_REAGENT_VOLUME[i] #* LST_PROTOCOL_IN_ML[i] 
            tmp_reag_wip_vol = prev_volume - deduct
           
            if tmp_reag_wip_vol < 0: 

                tmp_pack_units -= 1 
                if (tmp_pack_units == 0): 
                    tmp_pack_units += incr_packs # increase packs
                tmp_reag_wip_vol += tmp_pack_units * pack_size
                
                
            LST_PACK_SIZE.append(pack_size)
            LST_REAGENT_WORK_IN_PROGRESS_VOLUME.append(tmp_reag_wip_vol)
            LST_NUMBER_OF_PACK_UNITS.append(tmp_pack_units)
            LST_WORK_STOCK_VOLUME.append(pack_size * tmp_pack_units)
            prev_volume = tmp_reag_wip_vol
 
            
    
    df = pd.DataFrame(list(zip(LST_DATES,
                           LST_REAGENT_NAME,
                           LST_PACK_SIZE,
                           LST_NUMBER_OF_PACK_UNITS,
                           LST_REAGENT_WORK_IN_PROGRESS_VOLUME,
                           LST_QC_CONTROL_RUN, 
                           LST_QC_FAILURE,
                           LST_QC_FAILURE_REPEAT,
                           LST_QC_TOTAL_RUNS,
                           LST_TEST_RUNS,
                           LST_REPEATED_TEST_RUNS,
                           LST_REAGENT_WASTAGE,
                           LST_CALIBRATION_RUN_SCHEDULED,
                           LST_CALIBRATION_RUN_ON_FAILURE,
                           LST_CALIBRATION_RUN,
                           LST_TOTAL_REAGENT_VOLUME,
                           LST_PROTOCOL_IN_ML)), 
               columns =['date',
                         'reagent_name',
                         'pack_size',
                         'num_pack_units', 
                         'reagent_wip_volume', 
                         'qc_control_runs',
                         'qc_failure',
                         'qc_failure_repeat',
                         'qc_total_runs',
                         'test_runs',
                         'repeated_test_runs' ,
                         'reagent_wastage',
                         'calibration_runs_scheduled',
                         'calibration_runs_on_failure',
                         'total_calibration_runs',
                         'total_reagent_volume',
                         'protocol_in_ml']) 
    return(df)
```
